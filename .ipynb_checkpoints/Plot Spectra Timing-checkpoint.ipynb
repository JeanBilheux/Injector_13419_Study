```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import mpld3 # using mpld3 (mpld3.github.io) # installed using > pip install mpld3
mpld3.enable_notebook() # to turn it on (issues on Safari)
```

```python
spectra_file_name = '/Volumes/Macintosh HD 3/Users/j35/IPTS/IPTS-13419/Image007_Spectra.txt'
```

```python
#Load spectra
spectra_file = open(spectra_file_name, 'r')
time = []
counts = []
for line in spectra_file:
    line = line.strip()
    columns = line.split("\t")
    time.append(float(columns[0])* 1e3)
    counts.append(long(columns[1]))
```

```python
fig1 = plt.figure()
index_file = np.arange(len(counts))
plt.plot(index_file, time)
plt.title('Time vs index of file')
plt.xlabel('Index file')
plt.ylabel('Time (ms)')
#plt.show()
mpld3.display()
```

```python
#checking the time intervals
delta_time = []
for i in np.arange(len(counts)-1):
    d_t = time[i+1] - time[i]
    delta_time.append(d_t)
plt.plot(index_file[0:-1], delta_time)
plt.xlabel('Index file')
plt.ylabel('Delta Time (ms)')
plt.title('Acquisition frequency')
#plt.show()
mpld3.display()
```

```python
print(u"Delta time at index 500 is %.2f\u00b5s (frequencey of %.2fHz)" %(delta_time[499]*1000, 1/(time[499]/1e3)))
```

```python
print(u"Delta time at index 1000 is %.2f\u00b5s (frequencey of %.2fHz)" %(delta_time[999]*1000, 1/(time[999]/1e3)))
```

```python
# find first and last files running at 300Hz
index_start = 200
ref_delta_time = delta_time[index_start] - delta_time[index_start-1]
for index in np.arange(index_start, 0, -1):
    live_delta_time = delta_time[index] - delta_time[index-1]
    if abs(live_delta_time - ref_delta_time)> 0.0001:
        print 'First file running at 300Hz is at index %d' %index
        break

for index in np.arange(index_start, 900):
    live_delta_time = delta_time[index] - delta_time[index+1]
    if abs(live_delta_time - ref_delta_time) > 0.0001:
        print 'Last file running at 300Hz is at index %d' %index
        break
        
```
